#### Bag of Words


example :-

 - sen1 : "this is a great book"
 - sen2 : "the fit is great"
 - sen3 : "i love the shoes"
 - sen4 : "i love the book"


text vocab : i love the book this is a great fit shoes

great on the words it has seen on training .. but for unknown words it fails miserably
 

```python
class Category:
    BOOKS = "BOOKS"
    CLOTHING = "CLOTHING"

train_x = ["i love the book", "the fit is great","i love the shoes","this is a great book"]
train_y = [Category.BOOKS,Category.CLOTHING,Category.CLOTHING,Category.BOOKS]
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
vector = vectorizer.fit_transform(train_x)

print(vectorizer.get_feature_names_out())

print(vector.toarray())
```

```python
from sklearn import svm

clf_svm = svm.SVC(kernel="linear")
clf_svm.fit(vector, train_y)
```

```python
test_x = vectorizer.transform(["i like the book","shoes are alright"])

clf_svm.predict(test_x)
```

Stemming / lemmatization

- techniques to normalize text .. to redue redundency for simillar words ,here we reduce it to its canonical forms
- nltk lib has thease methods

check the fundamental difference between these 2

```python
import nltk

nltk.download("wordnet")
nltk.download("stopwords")
nltk.download("punkt_tab")
```

```python
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
phrase = "reading the books"

words = word_tokenize(phrase)

stemmed_words = []
for word in words:
    stemmed_words.append(stemmer.stem(word))

" ".join(stemmed_words)

# now we can build our word embeddings / vectors for this stemmed words .. which will give  good reasults 
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

phrase = "reading the books"
words = word_tokenize(phrase)

lemmatizer_words = []
for word in words:
    lemmatizer_words.append(lemmatizer.lemmatize(word))

" ".join(lemmatizer_words)
```

Stopwords Removal

- the a set of most common words in english (the, is , are ...) , these carry very less meaning in sentence vectors
so we remove them

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

stop_words = stopwords.words('english')

phrase = "her is an example sentence for removing stopwords , how do you like that methods"

words = word_tokenize(phrase)

stripped_phrases = []
for word in words:
    if word not in stop_words:
        stripped_phrases.append(word)

" ".join(stripped_phrases)
```
